# **Example**

```python
# import the necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load the image
image = cv2.imread('car.jpg')
```

```python
# show the image size
image.shape
```

```python
# show image
plt.imshow(image)
plt.show()
```

```python
# convert BGR to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# show image
plt.imshow(image_rgb)
plt.show()
```

```python
# show the value of a single pixel
image[350][250]
```

## Image Resizing

```python
# resize the image
new_width = 400
new_height = 100
little_image = cv2.resize(src=image_rgb, dsize=(new_width, new_height))

plt.imshow(little_image)
plt.show()
```

```python
# Resize the image with scaling

# Define the scale factor
# Decrease the size by 3 times
scale_factor = 1/3.0

# Get the original image dimensions
height, width = image_rgb.shape[:2]

# Calculate the new image dimensions
new_height = int(height * scale_factor)
new_width = int(width * scale_factor)

# Scaled image
scaled_image = cv2.resize(src= image_rgb, dsize =(new_width, new_height))

plt.imshow(scaled_image)
plt.show()
```

## Image Rotation

```python
# Image rotation parameter
center = (image_rgb.shape[1] // 2, image_rgb.shape[0] // 2)
angle = 30
scale = 1

# getRotationMatrix2D creates a matrix needed for transformation.
rotation_matrix = cv2.getRotationMatrix2D(center, angle, scale)

# We want matrix for rotation w.r.t center to 30 degree without scaling.
rotated_image = cv2.warpAffine(image_rgb, rotation_matrix, (image.shape[1], image.shape[0]))

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(7, 4))

# Plot the original image
axs[0].imshow(image_rgb)
axs[0].set_title('Original Image')

# Plot the Rotated image
axs[1].imshow(rotated_image)
axs[1].set_title('Image Rotation')

# Remove ticks from the subplots
for ax in axs:
    ax.set_xticks([])
    ax.set_yticks([])

# Display the subplots
plt.tight_layout()
plt.show()
```

```python
# 1. Tuliskan code untuk image translation seperti contoh di bawah

# read image using library skimage
image = io.imread("https://media.autoexpress.co.uk/image/private/s--vZkbi5D1--/f_auto,t_content-image-full-desktop@1/v1685458010/autoexpress/2023/05/Porsche%20911%20GTS%20UK%20001_otx6j7.jpg")

# translate image 200 pixels to the width and 53 pixels to the height as the output
translated_matrix = np.float32([[1, 0, 200], [0, 1, 53]])

# translated image ready
translated_image = cv2.warpAffine(image, translated_matrix, (image.shape[1], image.shape[0]))

# create subplots as example
fig, axs = plt.subplots(1, 2, figsize=(7, 4))

# set plot the original image
axs[0].imshow(image)
axs[0].set_title('Original Image')

# set plot the translated image
axs[1].imshow(translated_image)
axs[1].set_title('Image Translation')

# miscellaneous
# remove ticks from the subplots
for ax in axs:
    ax.set_xticks([])
    ax.set_yticks([])

# show
# display the subplots
plt.tight_layout()
plt.show()
```
